# Subject selction

```python
import pandas as pd
```

```python
df = pd.read_csv('hcp_unrestricted_data.csv')
dfr = pd.read_csv('hcp_restricted_data.csv')
```

### make sure we don't look at multiple subjects from the same family to avoid Oxford mafia

```python
assert len(sum(dfr.groupby(['Mother_ID', 'Father_ID']).groups.values(), [])) == 970
```

```python
unique_index = [x[-1] for x in dfr.groupby(['Mother_ID', 'Father_ID']).groups.values()]
```

### which are the subjects for which freesurfe outputs are missing on AWS?

```python
not_related_df = df.ix[unique_index]
not_related_df_final_subjects = not_related_df[not_related_df.MEG_AnyData]
freesurfer_missing = not_related_df_final_subjects.Subject[not_related_df_final_subjects.Release == 'S900'].astype(str).tolist()
```

```python
not_related_df_final_subjects
```

```python
print('|'.join(freesurfer_missing))
```

```python
missing_fsurfer = pd.DataFrame()
missing_fsurfer['SubjectId'] = not_related_df_final_subjects.Subject[not_related_df_final_subjects.Release == 'S900']
missing_fsurfer['CodedId'] = list(range(len(missing_fsurfer)))
missing_fsurfer.to_csv('missing_freesurfer_s900_unrelated.csv', index=False)
not_related_df_final_subjects.to_csv('unrelated_MEG_subjects.csv')
```
